```python
import requests

def search_product_prices(product_name):
    
    url = "https://serpapi.com/search.json"
    params = {
        "engine": "google_shopping",
        "q": product_name,
        "hl": "en",
        "gl": "in",
        "api_key": SERPAPI_API_KEY
    }

    response = requests.get(url, params=params)
    data = response.json()

    products = []

    for item in data.get("shopping_results", [])[:5]:
        products.append({
            "title": item.get("title"),
            "price": item.get("price"),
            "source": item.get("source"),
            "rating": item.get("rating"),
            "reviews": item.get("reviews"),
            "thumbnail": item.get("thumbnail", "N/A"),
        })

    return products
```

```python
from openai import OpenAI

client = OpenAI(
    api_key=GEMINI_API_KEY,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/",
)

def recommend_best_deal(products):
    prompt = f"""
You are an AI shopping agent.

Product price data:
{products}

Tasks:
1. Identify the cheapest product
2. Also give the comparison with other online Shooping App in tabular form.
3. Recommend the best deal
4. Explain the reason simply
"""

    response = client.chat.completions.create(
        model="gemini-2.5-flash",
        messages=[
            {"role": "system", "content": "You are a smart shopping assistant."},
            {"role": "user", "content": prompt},
        ],
        stream=True
    )
    
    full_response = "" 
    
    for chunk in response:
        delta = chunk.choices[0].delta
        if delta and delta.content:
            full_response += delta.content

    return full_response
```

```python
def run_agent():

    product = input("Enter product name: ")

    print("\nüîç Searching product prices...\n")
    products = search_product_prices(product)

    if not products:
        print("No products found.")
        return

    print("üì¶ Products Found:\n")
    for p in products:
        print(
            f"- {p['title']} | {p['price']} | {p['source']} | {p['rating']} | {p['reviews']} | {p['thumbnail']}"
        )

    print("\nü§ñ AI Recommendation:\n")
    decision = recommend_best_deal(products)
    print(decision)


if __name__ == "__main__":
    run_agent()
```
